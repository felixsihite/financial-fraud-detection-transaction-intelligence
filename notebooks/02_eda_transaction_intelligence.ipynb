```python
import json

import numpy as np
import pandas as pd
import plotly.express as px

from src.config import PROCESSED_DATA_DIR, REPORT_DIR
from src.visualization import APPROVED_PALETTE
```

```python
quality = json.loads((REPORT_DIR / "data_quality_summary.json").read_text(encoding="utf-8"))
type_summary = pd.DataFrame(quality["transaction_type_summary"])
type_summary["fraud_rate_pct"] = type_summary["fraud_rate"] * 100
type_summary
```

```python
px.bar(
    type_summary,
    x="type",
    y="transactions",
    text="transactions",
    title="Transaction Volume by Type",
    color_discrete_sequence=[APPROVED_PALETTE["accent_blue"]],
)
```

```python
px.bar(
    type_summary,
    x="type",
    y="fraud_rate_pct",
    text=type_summary["fraud_rate_pct"].map(lambda value: f"{value:.3f}%"),
    title="Fraud Rate by Transaction Type",
    color_discrete_sequence=[APPROVED_PALETTE["risk_red"]],
)
```
